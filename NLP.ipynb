```python
math_test = pd.read_csv('./math_test.csv', 'utf8', engine='python')
math_test['topic'] = pd.Series(1, index=math_test.index)
finance_test = pd.read_csv('./finance_test.csv', 'utf8', engine='python')
finance_test['topic'] = pd.Series(0, index=finance_test.index)
#테스트 셋 데이터 로드. 출력은 생략합니다!
math = math.T
finance = finance.T
math_test = math_test.T
finance_test = finance_test.T
#tokenize 이슈로 array 형태에 알맞게 조정
```

```python
from konlpy.tag import Twitter
pos_tagger = Twitter()

def tokenize(doc):
    # norm, stem은 optional
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]

math_data = [(tokenize((math[row])[0]), (math[row])[1]) for row in math]
finance_data = [(tokenize((finance[row])[0]), (finance[row])[1]) for row in finance]
print(math_data[0])
#수학의 첫번째 문서에 대해 토큰화한 결과를 도출.
#토큰화는 품사태깅 및 및 분절화를 시행합니다. 여기서는 Twitter 라이브러리의 pos 함수를 사용합니다.
```

```python
math_test_data = [(tokenize((math_test[row])[0]), (math_test[row])[1]) for row in math_test]
finance_test_data = [(tokenize((finance_test[row])[0]), (finance_test[row])[1]) for row in finance_test]
print(finance_test_data[0])
#금융의 첫번째 문서에 대해서도 같은 방법을 취해봅니다.
```

```python
math_data.extend(finance_data)
math_test_data.extend(finance_test_data)
# 학습을 위해 훈련/테스트 데이터 셋을 합칩니다.
tokens = [t for d in math_data for t in d[0]]
print(len(tokens))
#총 토큰의 갯수를 살펴봅니다.
```

```python
import nltk
text = nltk.Text(tokens, name='Train_Document_Token')
print(len(text.tokens))       #총 토큰의 갯수         
print(len(set(text.tokens)))  #유일한 토큰의 갯수 = (종류)         
pprint(text.vocab().most_common(50)) #상위 50번째 토큰까지 출력해봅니다.

# 관찰 : +, x 등 수학에 관련된 feature들도 높게 출력되지만 대부분의 상위 토큰은 "대다수의 문서에서 똑같이 많이 등장하는" 의미없는 토큰.
```

```python
selected_words = [f[0] for f in text.vocab().most_common(2000)]
selected_words1 = selected_words[1:2000]
selected_words2 = selected_words[100:2000]
selected_words3 = selected_words[500:2000]
def term_exists_V1(doc):
    return {'exists({})'.format(word): (word in set(doc)) for word in selected_words1}
def term_exists_V2(doc):
    return {'exists({})'.format(word): (word in set(doc)) for word in selected_words2}
def term_exists_V3(doc):
    return {'exists({})'.format(word): (word in set(doc)) for word in selected_words3}

train = [(term_exists_V1(d[0]), d[1]) for d in math_data]
test = [(term_exists_V1(d[0]), d[1]) for d in math_test_data]
# 나이브 베이즈 학습을 시행하기 위해 처리합니다.
```

```python
#빈도수 1위~2000위를 피쳐로 사용
classifier = nltk.NaiveBayesClassifier.train(train)
print("classification accuracy :" , nltk.classify.accuracy(classifier, test))
classifier.show_most_informative_features(10)
```

```python
#빈도수 100위~2000위를 피쳐로 사용
train = [(term_exists_V2(d[0]), d[1]) for d in math_data]
test = [(term_exists_V2(d[0]), d[1]) for d in math_test_data]
classifier = nltk.NaiveBayesClassifier.train(train)
print("classification accuracy :" , nltk.classify.accuracy(classifier, test))
classifier.show_most_informative_features(10)
```

```python
#빈도수 500위~2000위를 피쳐로 사용
train = [(term_exists_V3(d[0]), d[1]) for d in math_data]
test = [(term_exists_V3(d[0]), d[1]) for d in math_test_data]
classifier = nltk.NaiveBayesClassifier.train(train)
print("classification accuracy :" , nltk.classify.accuracy(classifier, test))
classifier.show_most_informative_features(10)

# 하위의 빈도수 단어를 포함할 수록 결과가 좋아진다는 것을 관찰할 수 있습니다.
# 그러나 모든 하위의 빈도수 단어를 포함하면 계속해서 결과가 좋아지는가? 는 생각해볼 문제입니다. 
# 첫번째로, 한두번밖에 등장하지 않았던 단어들도 포함될 여지가 있기 때문이고
# 두번째로, 아래 결과표와 같이 특징을 우리의 학습 데이터에만 잡을 여지가 있기 때문입니다. (2008, 2012, 2013과 같은 feature)
```

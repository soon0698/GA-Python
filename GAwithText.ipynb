## Library Import

```python
from random import (choice, random, randint)
import numpy as np
import matplotlib.pyplot as plt
from time import sleep
from GA import Population
%matplotlib inline
```

 ## 찾아갈 타겟 문자열 선정 및 최대 반복 횟수 설정

```python
_target  ="Hello, world! this is GA algorithm"
maxGenerations = 16384
```

## Population 파라미터 설정 

```python
pop = Population(size=2048, crossover=0.8, elitism=0.1, mutation=0.3)
```

## 타겟 문자열에 대한 Plot 그리기

```python
arr = np.arange(0, len(_target), 1)
s = np.zeros(len(_target))
for i in range(len(_target)):
    seq = ord(_target[i])
    s[i] = seq
plt.scatter(arr, s, s=60, c='red')
plt.show()
```

## 5개의 다른 Population들에 대하여 GA 알고리즘 실행.
### 60번까지만 반복하여 그래프 개형을 파악합니다.

```python
for i in range(1, maxGenerations + 1):
    plt.scatter(arr, s, s=60, c='red') 
    gene = np.zeros(len(_target)) # length : 34 , [0,0,0, ...]
    
    for j in range(5):
        for k in range(len(pop.population[j].invidual_fitness)):
            seq = ord(pop.population[j].gene[k])
            gene[k] =+ seq # ex. ascii value "abc" => [97, 98, 99, ...] 
            mask = gene > 1
            plt.scatter(arr[mask], gene[mask], alpha = 1 - (i /maxGenerations), marker= '+')
            
    if i % 20 == 0 and i <= 60:
        print("Generation %d: %s  | Label : %s" % (i, pop.population[0].gene, _target)) # 첫번째 population의 문자열만 관찰합니다.
        plt.show()
    if i > 60: break
    else:
        pop.evolve()
else:
    print("Maximum generations reached without success.")
```

### 실제 GA와 문자열 매칭 과정은 다음과 같이 이루어집니다. 

```python
for i in range(1, 400):
    if i % 20 == 0:
        print("Generation %d: %s  | Label : %s" % (i, pop.population[0].gene, _target))
    if pop.population[0].fitness == 0: break
    else:pop.evolve()
```
